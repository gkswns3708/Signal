```python
import json
from pprint import pprint
with open("./Training_Server/custom_dataset/deep.json", 'r') as f:
    json_value = json.load(f)
    
pprint(json_value)
```

```python
image_list = []
caption_list = []
for  idx, data in enumerate(json_value['dataset']):
    file_name = data['path'].split('/')[-1]
    # new_path = './custom_dataset/deep/' + file_name
    caption = data['caption']
    image_dict = {
        'file_name' : file_name,
        "id" : idx
    }
    caption_dict = {
        'image_id' : idx,
        'caption' : caption
    }
    image_list.append(image_dict)
    caption_list.append(caption_dict)
    
```

```python
dict_json = {
    'images' : image_list,
    'annotations' : caption_list
}

with open("./Training_Server/custom_dataset/custom_json.json", 'w') as f:
    json.dump(dict_json, f)
    
```
